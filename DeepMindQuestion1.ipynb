```python
#Import of libraries
import numpy as np
```

```python
#Initialization of a list for the number of days in a month for an ordinary year
EndMonth=np.array([31,28,31,30,31,30,31,31,30,31,30,31])
```

```python
#Cumulative sum for an ordinary year
CumEndMonth=np.cumsum(EndMonth)
CumEndMonth
```

```python
#Cumulative sum for a bissextile year
CumEndMonthBis= CumEndMonth+1
CumEndMonthBis[0]-=1
CumEndMonthBis
```

```python
#Initialization of variables for the year 2001 and the year 2100
start=2001
end=2101
```

```python
#Initialization of counters for last wednesday, all wednesday, the day number which corresponds to the first
#wednesday of 2001, i.e 3 as the year 2001 started a 01 Monday.
LastWedCtr=0
AllWedCtr=1
Wed=3

#Loop through all years
for year in range(start,end):
    #Initialize the list for the number of days in a month depending of the type of the current year
    EndMonth=[]
    if year%4==0:
        EndMonth=CumEndMonthBis
    else:
        EndMonth=CumEndMonth
    #Check if year is a century year which is not divisible by 400    
    if year%100==0 and year%400!=0:
        EndMonth=CumEndMonth
        
    #While the wednesday number for a given year is less than the total number of the days in that year,
    #increment the wednesday number by 7. Update the counter for the total number of wednesdays as well
    #as the one for the counter for the wednesdays which falls in the end of a month.

    while Wed <= EndMonth[11]-7:
        Wed+=7
        AllWedCtr+=1
        
        if Wed in EndMonth:
            LastWedCtr+=1
            print(year,Wed,np.where(EndMonth==Wed)[0]+1,"\n")
    
    #Reinitilize the counter for the day number of a wednesday according to the day of the last wednesday 
    #for the current year
    DayRest=EndMonth[11]-Wed
    Wed=7-DayRest
    
```

```python
#Print the total number of wednesdays and the total number of wednesdays which fall on the last day
#of a month
AllWedCtr,LastWedCtr
```

```python
#Compute the percentage of wednesdays which fall on the last day of a month for this century
perc=LastWedCtr/AllWedCtr*100
perc
```
